# Anomaly detection

## Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager
from sklearn import svm
from sklearn.datasets.samples_generator import make_blobs
from sklearn.neighbors import LocalOutlierFactor
```

## Generation of the network

```python
num_grupos=1
points, y_true = make_blobs(n_samples=300, centers=num_grupos,
                       cluster_std=0.70, random_state=2)
plt.scatter(points[:, 0], points[:, 1], s=50);
```

## Anomaly detection with svm

```python
# fit the model
clf = svm.OneClassSVM(nu=0.03, kernel="rbf", gamma=0.2)
clf.fit(points)
y_pred_train = clf.predict(points)
```

## Results of anomaly detection with svm


```python
plt.scatter(points[:, 0], points[:, 1], c=y_pred_train, s=50, alpha=0.5)
plt.show()
```

## Anomaly detection with neighbours

```python
# fit the model
clf2 = LocalOutlierFactor(n_neighbors=20, novelty=True, contamination=0.02)
clf2.fit(points)
y_pred_train2 = clf2.predict(points)
```

## Results of anomaly detection with neighbours


```python
plt.scatter(points[:, 0], points[:, 1], c=y_pred_train2, s=50, alpha=0.5)
plt.show()
```
