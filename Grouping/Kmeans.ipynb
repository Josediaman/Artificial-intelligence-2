# K-Means

## Packages

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
```

## Network

```python
num_grupos=6
points, y_true = make_blobs(n_samples=300, centers=num_grupos,
                       cluster_std=0.60, random_state=0)
plt.scatter(points[:, 0], points[:, 1], s=50);
```

## Model

```python
iteration_n = 100
error=[]
nc=range(2,10)

for i in nc:
  kmeans = KMeans(n_clusters=i, random_state=0).fit(points)
  error.append(kmeans.inertia_)
```

```python
plt.xlabel('Número de clústeres (k)')
plt.ylabel('Suma de los errores cuadráticos')
plt.plot(nc,error)
```

```python
# 5 parece ser el codo del brazo
kmeans_final = KMeans(n_clusters=6, random_state=0).fit(points)
```

## Results

```python
plt.scatter(points[:, 0], points[:, 1], c=kmeans_final.labels_, s=50, alpha=0.5)
plt.plot(kmeans_final.cluster_centers_[:,0], kmeans_final.cluster_centers_[:,1], 'kx', markersize=15)
plt.show()
```
