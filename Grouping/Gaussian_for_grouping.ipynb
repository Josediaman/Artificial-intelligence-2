# Gaussian for grouping

## Packages

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.datasets.samples_generator import make_blobs
```

## Network

```python
num_grupos=6
points, y_true = make_blobs(n_samples=300, centers=num_grupos,
                       cluster_std=0.60, random_state=0)
plt.scatter(points[:, 0], points[:, 1], s=50);
```

## Model

```python
score=[]
nc=range(2,10)

for i in nc:
  gm = GaussianMixture(n_components=i, random_state=0).fit(points)
  scor=gm.score(points)
  score.append(scor)

plt.xlabel('Número de clústeres (k)')
plt.ylabel('Suma de los errores cuadráticos')
plt.plot(nc,score)
```

```python
# 6 parece ser el codo del brazo 
gm = GaussianMixture(n_components=6, random_state=0).fit(points)
labels=gm.predict(points)
```

## Results

```python
plt.scatter(points[:, 0], points[:, 1], c=labels, s=50, alpha=0.5)
plt.show()
```
